```python
def rngrams(input, n, output):
    input = input.split(' ')
    for i in range(len(input)-n+1):
        if(input[i] == input[i+1]):
            g = ' '.join(input[i:i+n])
            output.setdefault(g, 0)
            output[g] += 1
    return output
```
