```python
 import re
dic={}
def rngrams(input):
    input = input.replace('\n',' ')
    input = input.replace('.',' ')
    input = input.replace(' ',' ')
    input = input.replace(',',' ')
    input = input.replace('،',' ')
    input = re.split(' ',input)
    
    for word in input:
            dic.setdefault(word, 0)
            dic[word] += 1
    
```

```python
for i in range(0,df.shape[0]):
    row = data.iloc[i]
    rngrams(row.desc)
    rngrams(row.title)
```
