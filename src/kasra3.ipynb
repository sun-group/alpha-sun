```python
df = data
```

```python
 import re
dic={}
dictCat = [] 
cats = list(data.cat2.unique())

for i in range(len(cats)):
    dictCat.append(dict())

def rngrams(input,catnum):
    input = input.replace('\n',' ')
    input = input.replace('.',' ')
    input = input.replace(' ',' ')
    input = input.replace(',',' ')
    input = input.replace('،',' ')
    input = re.split(' ',input)
    
    for word in input:
        dic.setdefault(word, 0)
        dic[word] += 1
        dictCat[catnum].setdefault(word,0)
        dictCat[catnum][word]+=1
    
```

```python
for i in range(df.shape[0]):
    row = data.iloc[i]
    rngrams(row.desc,cats.index(row["cat2"]))
    #rngrams(row.title)
    if (i%1000 == 0):
        print(i)
```

```python
descsInCats = [0]*len(cats)
for i in range(len(cats)):
    for w in dictCat[i].keys():
        descsInCats[i]+=dictCat[i].get(w)
```

```python
word_list = list(dic.keys())
word_list.remove("")
```

```python
def probab(index , word):
    if (word in dictCat[index].keys()):
        count = dictCat[index].get(word)
    else:
        count = 0
    total = descsInCats[index]
    return count/total
```

```python
goods = set()
var = {}
for w in word_list:
    sump = 0
    p = [0]*len(cats)
    for i in range(len(cats)):
        p[i] = probab(i,w)
        sump += p[i]
    for i in range(len(cats)):
        if (p[i]==0):
            continue
        p[i]/=sump
    var[w] = stats.entropy(p)
```
