```python
def counter(input, n, string):
    output = dict()
    input = input.replace(' ','ـ')
    input = input.replace('.','ـ')
    input = input.replace('-','ـ')
    input = input.replace('،','ـ')
    input = re.split('ـ',input)
    for i in range(len(input)-n+1):
        if(input[i] == string):
            for j in range(1,n):
                output.setdefault(input[i+j], 0)
                output.update({input[i+j]:output.get(input[i+j])+1})
    return output
```

```python
d = dict()
goods = ['تبریز','بازدید','محل','شهرستان','در','شهر','موقعیت','تحویل','طرفهای','جاده']
for s in tab:
    for i in range(2,4):
        for q in goods:
            d.update(counter(s,i,q))
```
